```python
import urllib.request 
import bs4
```

```python
url2 = "https://news.naver.com/"

html2 = urllib.request.urlopen(url2)
```

```python
print(html2)
```

```python
bs_obj2 = bs4.BeautifulSoup(html2.read(), "html.parser")
```

```python
print(bs_obj2)
```

```python
ul2 = bs_obj2.find("ul", {"id":"text_today_main_news_801001"})
```

```python
ul2
```

```python
lis2 = ul2.findAll("li")
```

```python
for li in lis2:
    print(li)
```
