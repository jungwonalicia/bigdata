```python
import requests
from bs4 import BeautifulSoup 
```

```python
def get_price2(company_code):
    url = "https://finance.naver.com/item/main.nhn?code=" + company_code
   #url = "https://finance.naver.com/item/main.nhn?code=005930"
    
    document = requests.get(url)
    html = BeautifulSoup(document.content, "html.parser")
      
    td_first = html.find("td", {"class": "first"})
     
    close_blind = td_first.find("span", {"class":"blind"})
      
   
    return  close_blind.text

```

```python
samsungElec = "005930"
hynix = "000660"
samjung = "005680"
company_codes2 = [samsungElec, hynix, samjung] #회사 리스트
```

```python
for item in company_codes2:
    price = get_price2(item)
    print("close price: ", item, price)
```

```python
company_list = ["삼성전자", "하이닉스", "삼정전자"]
```

```python
i = 0
for item in company_codes2:
    price = get_price2(item)
    print("close price: ", company_list[i], item, price)
    i = i + 1
```

```python
def get_price3(company_code):
    url = "https://finance.naver.com/item/main.nhn?code=" + company_code
   #url = "https://finance.naver.com/item/main.nhn?code=005930"
    
    document = requests.get(url)
    html = BeautifulSoup(document.content, "html.parser")
      
    td_first = html.find("td", {"class": "first"})
    close_blind = td_first.find("span", {"class":"blind"})
    close_text = close_blind.text
    
    high_text = ""
   
    return  {"close":close_text, "high":high_text} #딕셔너리 형태로 값을 여러개 넘길 수 있다.
```

```python
i = 0
for item in company_codes2:
    price = get_price3(item)
    print("close & high price: ", company_list[i], item, price)
    i = i + 1
```

```python
def get_price4(company_code):
    url = "https://finance.naver.com/item/main.nhn?code=" + company_code
   #url = "https://finance.naver.com/item/main.nhn?code=005930"
    
    document = requests.get(url)
    html = BeautifulSoup(document.content, "html.parser")
      
    td_first = html.find("td", {"class": "first"})
    close_blind = td_first.find("span", {"class":"blind"})
    close_text = close_blind.text
    
    table = html.find("table", {"class":"no_info"})
    trs = table.findAll("tr")
    print(trs)
    #소스보기에서는 안보이지만, 생성된 코드를 보면 원하는 값이 어디에 있는지 확인할 수 있다.
    
    first_tr = trs[0]; #첫번째 tr
    first_tr_tds = first_tr.findAll("td")
    print(first_tr_tds) #결과가 여러개인 경우 리스트 형태로 리턴.
                        #생성된 html태그를 보고, 원하는 값의 위치를 찾아낸다.
    first_tr_tds_second_td = first_tr_tds[1]
    print(first_tr_tds_second_td)
    
    high = first_tr_tds_second_td.find("span", {"class":"blind"})
    
    high_text = high.text
   
    return  {"close":close_text, "high":high_text} #딕셔너리 형태로 값을 여러개 넘길 수 있다.
```
